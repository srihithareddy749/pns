```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

study_hours = np.array([2, 4, 6, 8, 1, 5, 7, 3, 9, 10])
exam_scores = np.array([50, 60, 70, 85, 40, 65, 78, 55, 90, 95])
sleep_hours = np.array([8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3])

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(study_hours, exam_scores, color='blue')
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study Hours vs. Exam Scores")

plt.subplot(1, 3, 2)
plt.scatter(sleep_hours, exam_scores, color='green')
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Scores")
plt.title("Sleep Hours vs. Exam Scores")

plt.subplot(1, 3, 3)
plt.scatter(study_hours, sleep_hours, color='red')
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study Hours vs. Sleep Hours")

plt.tight_layout()
plt.show()

corr_study_exam, _ = pearsonr(study_hours, exam_scores)
corr_sleep_exam, _ = pearsonr(sleep_hours, exam_scores)
corr_study_sleep, _ = pearsonr(study_hours, sleep_hours)

print(f"Correlation between Study Hours and Exam Scores: {corr_study_exam:.2f}")
print(f"Correlation between Sleep Hours and Exam Scores: {corr_sleep_exam:.2f}")
print(f"Correlation between Study Hours and Sleep Hours: {corr_study_sleep:.2f}")

if corr_study_exam > 0:
    print("There is a positive correlation between study hours and exam scores, meaning more study hours generally lead to higher scores.")
if corr_sleep_exam < 0:
    print("There is a negative correlation between sleep hours and exam scores, meaning less sleep is associated with higher scores.")
if corr_study_sleep < 0:
    print("There is a negative correlation between study hours and sleep hours, meaning more study hours tend to reduce sleep hours.")
```
