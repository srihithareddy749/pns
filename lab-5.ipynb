```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data (number of vehicles per household for N households)
N = 1000  # Number of households
np.random.seed(42)
vehicles_per_household = np.random.choice([0, 1, 2, 3, 4, 5], size=N, p=[0.1, 0.3, 0.4, 0.1, 0.07, 0.03])

# Compute PMF
values, counts = np.unique(vehicles_per_household, return_counts=True)
pmf = counts / N

# Compute CDF
cdf = np.cumsum(pmf)

# Plot PMF
plt.figure(figsize=(10, 5))
sns.barplot(x=values, y=pmf, color='blue', alpha=0.6)
plt.xlabel("Number of Vehicles per Household")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")
plt.grid(True)
plt.show()

# Plot CDF
plt.figure(figsize=(10, 5))
sns.lineplot(x=values, y=cdf, marker="o", color='red')
plt.xlabel("Number of Vehicles per Household")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats

# Given values
N = 100  # Number of inspected gadgets
p = 0.05  # Probability of a gadget being defective

# i) Probability of exactly 10 defective gadgets
prob_10 = stats.binom.pmf(10, N, p)

# ii) Probability of at least 15 defective gadgets
prob_at_least_15 = 1 - stats.binom.cdf(14, N, p)

# iii) Expected number of defective gadgets
expected_defective = N * p

print(f"i) P(X = 10) = {prob_10:.4f}")
print(f"ii) P(X ≥ 15) = {prob_at_least_15:.4f}")
print(f"iii) Expected defective gadgets = {expected_defective}")
```
